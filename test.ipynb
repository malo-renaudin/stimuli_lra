```python
import pandas as pd
import numpy as np

def find_optimal_balancing_thresholds(df, min_words_per_bin=1):
    """
    Finds the optimal syllable and frequency thresholds.
    
    Prioritizes:
    1. Minimizing the number of empty bins (maximizing total active bins out of 16).
    2. Maximizing the minimum number of words in the tightest bin (as a tie-breaker).
    """
    best_syll_threshold = None
    best_freq_threshold = None
    max_active_bins = -1
    max_min_count = -1
    best_grid_summary = {}

    # 1. Test possible syllable split values
    possible_syll_cuts = sorted(df["Lexique4__SyllNb"].dropna().unique())
    if len(possible_syll_cuts) > 1:
        possible_syll_cuts = possible_syll_cuts[:-1]

    # 2. Test different frequency percentiles (from the 5th to the 90th percentile)
    possible_freq_percentiles = np.linspace(5, 90, 200) 

    print("Scanning data distribution matrix...")

    for syll_cut in possible_syll_cuts:
        len_cond = df["Lexique4__SyllNb"].apply(lambda x: "long" if x > syll_cut else "short")
        
        for pct in possible_freq_percentiles:
            freq_cut = np.percentile(df["Lexique4__FreqOrtho"], pct)
            freq_cond = df["Lexique4__FreqOrtho"].apply(lambda x: "high" if x > freq_cut else "low")
            
            current_min_bin_count = float('inf')
            active_bins_count = 0
            temp_counts = {}
            
            # Cross-examine all 16 bins
            for f in ["high", "low"]:
                for l in ["long", "short"]:
                    for g in df["gender"].unique():
                        for n in df["number"].unique():
                            mask = (freq_cond == f) & (len_cond == l) & (df["gender"] == g) & (df["number"] == n)
                            count = mask.sum()
                            
                            temp_counts[f"{f}_{l}_{g}_{n}"] = count
                            
                            if count > 0:
                                active_bins_count += 1
                                
                            if count < current_min_bin_count:
                                current_min_bin_count = count
            
            # Condition 1: This setup gives us strictly MORE populated bins
            # Condition 2: Equal number of populated bins, but a healthier bottleneck (tie-breaker)
            if (active_bins_count > max_active_bins) or \
               (active_bins_count == max_active_bins and current_min_bin_count > max_min_count):
                
                max_active_bins = active_bins_count
                max_min_count = current_min_bin_count
                best_syll_threshold = syll_cut
                best_freq_threshold = freq_cut
                best_grid_summary = temp_counts

    # 3. Output results
    empty_bins_count = 16 - max_active_bins
    
    if empty_bins_count == 0 and max_min_count >= min_words_per_bin:
        print(f"\n🎉 Success! Found optimal operational thresholds:")
        print(f"-> Syllable Threshold: Words with > {best_syll_threshold} syllables are 'long' (Short: <= {best_syll_threshold})")
        print(f"-> Frequency Threshold: Words with > {best_freq_threshold:.4f} per million are 'high'")
        print(f"-> All 16 bins are filled! Every bin has at least **{max_min_count}** candidate word(s).\n")
    else:
        print(f"\n❌ Warning: Could not fill all 16 bins evenly with the current data pool.")
        print(f"-> Best Syllable Threshold Attempt: > {best_syll_threshold} syllables are 'long'")
        print(f"-> Best Frequency Threshold Attempt: > {best_freq_threshold:.4f} per million are 'high'")
        print(f"The best configuration still left **{empty_bins_count}** empty categories (Active bins: {max_active_bins}/16).")
        print("See the matrix breakdown below to find which categories need more words:\n")
        
    # Display the breakdown cleanly
    print("--- Best Attempt Matrix Distribution ---")
    for bin_name, count in best_grid_summary.items():
        alert = " ⚠️ EMPTY BIN" if count == 0 else ""
        print(f"Bin [{bin_name.replace('_', ' ')}]: {count} words available{alert}")
        
    return best_freq_threshold, best_syll_threshold
```
